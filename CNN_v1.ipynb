```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Training data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)

training_set = train_datagen.flow_from_directory(
    '/home/arshia/Downloads/archive/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Validation data (no augmentation, only rescaling)
valid_datagen = ImageDataGenerator(rescale=1./255)

validation_set = valid_datagen.flow_from_directory(
    '/home/arshia/Downloads/archive/val',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)


# Test data (no augmentation, no shuffle!)
test_datagen = ImageDataGenerator(rescale=1./255)

test_set = test_datagen.flow_from_directory(
    '/home/arshia/Downloads/archive/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)
```

```python
cnn = tf.keras.models.Sequential()

# First Convolution + Pooling
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(224, 224, 3)))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))

# Flatten
cnn.add(tf.keras.layers.Flatten())

# Fully connected layers
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))

# Dropout to reduce overfitting
cnn.add(tf.keras.layers.Dropout(0.5))

# Output layer
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = cnn.fit(training_set, validation_data=validation_set, epochs=5)
```

```python
test_loss, test_acc = cnn.evaluate(test_set)
print("Test accuracy:", test_acc)
```
